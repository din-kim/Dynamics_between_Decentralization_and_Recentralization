```python
# Import Classes
import random
from Reality import *
from Organization import *
from User import *
from functions import *

# Import libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
np.set_printoptions(4)
random.seed(11)
```

```python
"""
Variables
- m : number of attributes
- n : number of users
- p : participation rate
- k : degree of interdependence
- t : total number of tokens
- dr: distribution rate of tokens
"""

rds = 10
v = 10
m = 100
n = 100
k = 0
t = 1000
dr = 0.2
tokens = list(distribute_tokens(n, t, dr))
ids = list(range(n))
```

```python
# Initiate Reality and Organization
reality = Reality(m)
organization = Organization(m, reality)

# Initiate Users
users = []
for i in range(n):
    globals()['user{}'.format(i)] = User(m, k, organization, ids, tokens)
    users.append(globals()['user{}'.format(i)])
```

```python
vots = []
deles = []
vot_anys = []
p_ys =[]
p_ns = []
knws = []
prfs = []

# Initiate Vote
for rd in range(rds):
    vote_list = generate_vote_list(m, v)
    vot, dele, vot_any, p_y, p_n, knw, prf = vote_handler(reality, organization, users, vote_list)
    vots.append(vot)
    deles.append(dele)
    vot_anys.append(vot_any)
    p_ys.append(p_y)
    p_ns.append(p_n)
    knws.append(knw)
    prfs.append(prf)
```
